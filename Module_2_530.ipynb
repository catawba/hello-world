**Section 1.4 #5 Exact ODE**

Solve

$$(x^2+y^2)dx - (2xy)dy = 0$$

Check for exact form
$$\frac{\partial P}{\partial y} = (x^2+2y)$$
$$\frac{\partial Q}{\partial x}= -(2y+2x)$$

Exactness test fails, so try multiplying by $$\frac{1}{x^2}$$

$$\frac{1}{x^2}((x^2+y^2)dx - (2xy)dy)=0$$

$$\frac{x^2+y^2}{x^2}dx - \frac{2xy}{x^2} dy=0$$

$$1+\frac{y^2}{x^2}dx - \frac{2y}{x}dy = 0$$

$$P(x)=1+\frac{y^2}{x^2}$$  $$Q(x) = -\frac{2y}{x}$$

$$\frac{\partial P}{\partial y} = \frac{2y}{x^2} and \frac{\partial Q}{\partial x} = \frac{2y}{x^2}$$

So we now have an exact form. 
let $$u=\int P(x)dx + k(y)$$

$$u = \int (1+\frac{y^2}{x^2})dx+k(y)$$

$$u = x+\frac{-y^2}{x}+ k(y)$$

now 

$$\frac{\partial u}{\partial y} = Q(x)$$

$$0+\frac{-2y}{x}+k'(y)=\frac{-2y}{x}$$

$$k'(y)=0$$

$$\int {k'(y)}=\int {0}$$

$$k(y)=c_1$$

$$u=0$$

$$x-\frac{y^2}{x}+c_1=0$$

Solution
$$x-\frac{y^2}{x}=c$$







**Section 1.4 #6  Exact ODE**

Solve
$$3(y+1)dx=2x dy$$ 
with integrating factor $$(y+1)x^{-4}$$

Check for exactness

$$3(y+1)dx - 2x dy = 0$$

Let
$$P(x,y)=3(y+1)$$ and
$$Q(x,y)=-2x$$

$$\frac{\partial{P}}{\partial{y}}=3$$
$$\frac{\partial{Q}}{\partial{x}}=-2$$

$$\frac{y+1)}{x^{4}}(3(y+1)dx - 2x dy)=0$$

$$\frac{3(y+1)^2}{x^4}dx-\frac{(y+1)(2x)}{x^4}dy=0$$

$$\frac{3(y+1)^2}{x^4}dx-\frac{2(y+1)}{x^3}dy$$
now $$P(x,y)=\frac{3(y+1)^2}{x^4}dx$$
$$Q(x,y)=-\frac{2(y+1)}{x^3}dy$$
giving
$$\frac{\partial{P}}{\partial{y}}=\frac{6(y+1)}{x^4}$$
$$\frac{\partial{Q}}{\partial{x}}=\frac{6(y+1)}{x^4}$$
we now have exactness, so we use
$$u=\int{M(x)}dx+k(y)$$
with 
$$u=\int{\frac{3(y+1)^2}{x^4}}dx+k(y)$$

$$u=\frac{-(y+1)^2}{x^3} + k(y)$$

now we set the partial of u with respect to y equal to Q(x,y) (The one we got after multiplying by the integrating factor)
$$\frac{\partial{u}}{\partial{y}}=Q(x,y)$$

$$\frac{-2(y+1)}{x^3}+\frac{\partial{k}}{\partial{y}}=\frac{-2(y+1)}{x^3}$$

Simplify
$$\frac{\partial{k}}{\partial{y}}=0$$
